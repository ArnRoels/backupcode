```python
import os
import glob
import cv2
import numpy as np
from skimage.segmentation import find_boundaries
from tifffile import imread, imwrite

# 1. Set your paths
input_dir = "/media/arnout/Elements1/groteEdeepprofilerdingen/deepprofileroutputsetc/testset_masking/inputs/outlines (Copy)"  # Where your current .tif masks are
output_dir = "/media/arnout/Elements1/groteEdeepprofilerdingen/deepprofileroutputsetc/testset_masking/inputs/outlines" # Where the processed outlines will go

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 2. Get all .tif mask files
mask_files = glob.glob(os.path.join(input_dir, "*.tif"))

print(f"Found {len(mask_files)} masks. Starting conversion...")

for mask_path in mask_files:
    # Load the Omnipose label mask
    # Omnipose labels usually have unique integers for each cell
    mask = imread(mask_path)
    
    # Generate boundaries: find_boundaries returns a boolean array 
    # where True indicates the edge of a labeled object
    outlines = find_boundaries(mask, mode='thick')
    
    # Convert boolean to uint8 (0 and 255) for DeepProfiler
    outlines_binary = (outlines.astype(np.uint8)) * 255
    
    # Create the output filename
    file_name = os.path.basename(mask_path)
    save_path = os.path.join(output_dir, file_name)
    
    # Save as .tif (or .png if you prefer)
    imwrite(save_path, outlines_binary, compression='zlib')

print("Done! Check the 'outlines_binary' folder.")
```

```python
import os
import shutil
import re

# --- CONFIGURATION ---
root_dir = "/media/arnout/Elements1/Thesis/Screen crispri quality checked"
outlines_base = "/media/arnout/Elements1/Thesis/project_masking_label_T2/inputs/outlines"

# Regex to find the mask files
mask_pattern = re.compile(r"Sample_[A-Z][0-9]+_XY[0-9]+_C1_cp_masks\.tif")

# Plates to process
plates = ["PLATE2_T0", "PLATE2_T1", "PLATE2_T2"]

transfer_count = 0
skipped_count = 0

for plate in plates:
    # 1. Setup the flat destination folder for this plate (e.g., outlines/PLATE3_T0)
    plate_outlines_dir = os.path.join(outlines_base, plate)
    os.makedirs(plate_outlines_dir, exist_ok=True)
    
    plate_path = os.path.join(root_dir, plate)
    if not os.path.exists(plate_path):
        print(f"Warning: Plate path {plate_path} does not exist. Skipping.")
        continue
    
    print(f"Processing {plate}...")

    # 2. Walk through the subfolders (Sample_A1, Sample_A2, etc.)
    for well_folder in os.listdir(plate_path):
        # Skip non-sample folders or quality-flagged folders
        if not well_folder.startswith("Sample_") or "bad_Q" in well_folder:
            continue
            
        mask_folder = os.path.join(plate_path, well_folder, "masks")
        
        if os.path.exists(mask_folder):
            for mask_name in os.listdir(mask_folder):
                # Verify it matches the mask filename pattern
                if mask_pattern.search(mask_name):
                    
                    source_path = os.path.join(mask_folder, mask_name)
                    # Use the original mask_name for the destination
                    dest_path = os.path.join(plate_outlines_dir, mask_name)
                    
                    # 3. Perform the copy
                    if not os.path.exists(dest_path):
                        shutil.copy2(source_path, dest_path)
                        transfer_count += 1
                    else:
                        skipped_count += 1

print(f"\n--- Process Complete ---")
print(f"Files copied to new folders: {transfer_count}")
print(f"Files already present (skipped): {skipped_count}")
print(f"Original masks are still safe in {root_dir}")
```

```python
import pandas as pd
import os
import re

# --- CONFIGURATION ---
# The base directory where your new PLATE folders (PLATE3_T0, etc.) are located
outline_base_path = "/media/arnout/Elements1/Thesis/project_masking_label_T2/inputs/outlines"
output_csv_path = "/media/arnout/Elements1/Thesis/project_masking_label_T2/inputs/metadata/outlines.csv"

# Regex to extract Well and Site from 'Sample_A1_XY1_C1_cp_masks.tif'
# This is more robust than splitting by underscores
name_pattern = re.compile(r"Sample_(?P<well>[A-Z][0-9]+)_XY(?P<site>[0-9]+)")

data = []

# Ensure metadata directory exists
os.makedirs(os.path.dirname(output_csv_path), exist_ok=True)

# 1. Iterate through each plate folder in the outlines directory
for plate_name in os.listdir(outline_base_path):
    plate_dir = os.path.join(outline_base_path, plate_name)
    
    # Skip if it's not a directory
    if not os.path.isdir(plate_dir):
        continue
        
    print(f"Adding {plate_name} to metadata...")

    # 2. Iterate through all .tif files in that plate folder
    for mask_name in os.listdir(plate_dir):
        if not mask_name.endswith(".tif"):
            continue
            
        match = name_pattern.search(mask_name)
        if match:
            well = match.group('well')
            site = match.group('site')
            
            # 3. Format the Outlines column as /PlateName/FileName
            relative_path = f"{plate_name}/{mask_name}"
            
            data.append([plate_name, well, site, relative_path])

# 4. Create DataFrame and Save
df = pd.DataFrame(data, columns=['Metadata_Plate', 'Metadata_Well', 'Metadata_Site', 'Outlines'])
df.to_csv(output_csv_path, index=False)

print(f"\nSuccess! outlines.csv created at: {output_csv_path}")
print(f"Total entries: {len(df)}")
```
